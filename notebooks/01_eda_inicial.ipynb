```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
```

```python

# Configurar estilo
sns.set_style("whitegrid")
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (20, 12)
plt.rcParams['font.size'] = 10

print("=" * 100)
print("ANÁLISIS EXPLORATORIO DE DATOS - FANTASY FOOTBALL 2024")
print("=" * 100)

# ============================================
# 1. ANÁLISIS DE DEFENSAS (DST)
# ============================================
print("\n" + "=" * 100)
print("1. ANÁLISIS DE DEFENSAS (DST)")
print("=" * 100)

print(f"\nTotal de equipos: {len(df_DST)}")
print(f"\nEstadísticas descriptivas:")
print(df_DST.describe())

print(f"\n--- Top 10 Defensas por FPTS ---")
top_dst = df_DST.nlargest(10, 'FPTS')
print(top_dst[['Rank', 'Player', 'SACK', 'INT', 'FR', 'DEF TD', 'SPC TD', 'FPTS', 'FPTS/G']])

# Gráficas DST
fig = plt.figure(figsize=(20, 12))
gs = GridSpec(3, 3, figure=fig)

# 1.1 Top 10 DST por FPTS
ax1 = fig.add_subplot(gs[0, :2])
top_dst_plot = df_DST.nlargest(10, 'FPTS')
ax1.barh(top_dst_plot['Player'], top_dst_plot['FPTS'], color='darkblue')
ax1.set_xlabel('Fantasy Points', fontsize=12, fontweight='bold')
ax1.set_title('Top 10 Defensas por Puntos Fantasy', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, v in enumerate(top_dst_plot['FPTS']):
    ax1.text(v + 0.5, i, f'{v:.1f}', va='center', fontweight='bold')

# 1.2 Distribución de FPTS/G
ax2 = fig.add_subplot(gs[0, 2])
ax2.hist(df_DST['FPTS/G'], bins=15, color='steelblue', edgecolor='black', alpha=0.7)
ax2.set_xlabel('FPTS/G', fontsize=10, fontweight='bold')
ax2.set_ylabel('Frecuencia', fontsize=10, fontweight='bold')
ax2.set_title('Distribución de FPTS por Juego', fontsize=12, fontweight='bold')
ax2.axvline(df_DST['FPTS/G'].mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {df_DST["FPTS/G"].mean():.2f}')
ax2.legend()

# 1.3 Correlación entre estadísticas defensivas
ax3 = fig.add_subplot(gs[1, :])
corr_cols = ['SACK', 'INT', 'FR', 'FF', 'DEF TD', 'SFTY', 'SPC TD', 'FPTS']
corr_matrix = df_DST[corr_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, ax=ax3, cbar_kws={'label': 'Correlación'})
ax3.set_title('Matriz de Correlación - Estadísticas Defensivas', fontsize=14, fontweight='bold')

# 1.4 Comparación de estadísticas defensivas (Top 10)
ax4 = fig.add_subplot(gs[2, :])
top_10_dst = df_DST.nlargest(10, 'FPTS')
x = np.arange(len(top_10_dst))
width = 0.15
ax4.bar(x - 2*width, top_10_dst['SACK'], width, label='Sacks', alpha=0.8)
ax4.bar(x - width, top_10_dst['INT'], width, label='Intercepciones', alpha=0.8)
ax4.bar(x, top_10_dst['FR'], width, label='Fumbles Rec', alpha=0.8)
ax4.bar(x + width, top_10_dst['DEF TD'], width, label='DEF TD', alpha=0.8)
ax4.bar(x + 2*width, top_10_dst['SPC TD'], width, label='SPC TD', alpha=0.8)
ax4.set_xlabel('Equipos', fontsize=12, fontweight='bold')
ax4.set_ylabel('Cantidad', fontsize=12, fontweight='bold')
ax4.set_title('Comparación de Estadísticas Defensivas - Top 10', fontsize=14, fontweight='bold')
ax4.set_xticks(x)
ax4.set_xticklabels([p.split('(')[0].strip()[:10] for p in top_10_dst['Player']], rotation=45, ha='right')
ax4.legend()
ax4.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('EDA_DST.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Gráficas de DST guardadas en 'EDA_DST.png'")
```

```python

# ============================================
# 7. ANÁLISIS COMPARATIVO ENTRE POSICIONES
# ============================================
print("\n" + "=" * 100)
print("7. ANÁLISIS COMPARATIVO ENTRE POSICIONES")
print("=" * 100)

fig = plt.figure(figsize=(20, 12))
gs = GridSpec(3, 2, figure=fig)

# 7.1 Comparación de FPTS promedio por posición
ax1 = fig.add_subplot(gs[0, 0])
positions = ['DST', 'K', 'QB', 'RB', 'TE', 'WR']
avg_fpts = [df_DST['FPTS'].mean(), df_K['FPTS'].mean(), df_QB['FPTS'].mean(), 
            df_RB['FPTS'].mean(), df_TE['FPTS'].mean(), df_WR['FPTS'].mean()]
colors_pos = ['darkblue', 'green', 'purple', 'orangered', 'teal', 'crimson']
ax1.bar(positions, avg_fpts, color=colors_pos, alpha=0.7, edgecolor='black')
ax1.set_ylabel('FPTS Promedio', fontsize=12, fontweight='bold')
ax1.set_title('Promedio de Puntos Fantasy por Posición', fontsize=14, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)
for i, v in enumerate(avg_fpts):
    ax1.text(i, v + 1, f'{v:.1f}', ha='center', fontweight='bold')

# 7.2 Boxplot de FPTS por posición
ax2 = fig.add_subplot(gs[0, 1])
fpts_data = [df_DST['FPTS'], df_K['FPTS'], df_QB['FPTS'], 
             df_RB['FPTS'], df_TE['FPTS'], df_WR['FPTS']]
bp = ax2.boxplot(fpts_data, labels=positions, patch_artist=True)
for patch, color in zip(bp['boxes'], colors_pos):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
ax2.set_ylabel('FPTS', fontsize=12, fontweight='bold')
ax2.set_title('Distribución de FPTS por Posición', fontsize=14, fontweight='bold')
ax2.grid(axis='y', alpha=0.3)

# 7.3 Número de jugadores por posición
ax3 = fig.add_subplot(gs[1, 0])
player_counts = [len(df_DST), len(df_K), len(df_QB), len(df_RB), len(df_TE), len(df_WR)]
ax3.bar(positions, player_counts, color=colors_pos, alpha=0.7, edgecolor='black')
ax3.set_ylabel('Número de Jugadores', fontsize=12, fontweight='bold')
ax3.set_title('Cantidad de Jugadores por Posición', fontsize=14, fontweight='bold')
ax3.grid(axis='y', alpha=0.3)
for i, v in enumerate(player_counts):
    ax3.text(i, v + 3, f'{v}', ha='center', fontweight='bold')

# 7.4 FPTS/G promedio por posición
ax4 = fig.add_subplot(gs[1, 1])
avg_fpts_per_game = [df_DST['FPTS/G'].mean(), df_K['FPTS/G'].mean(), df_QB['FPTS/G'].mean(), 
                     df_RB['FPTS/G'].mean(), df_TE['FPTS/G'].mean(), df_WR['FPTS/G'].mean()]
ax4.bar(positions, avg_fpts_per_game, color=colors_pos, alpha=0.7, edgecolor='black')
ax4.set_ylabel('FPTS/G Promedio', fontsize=12, fontweight='bold')
ax4.set_title('Promedio de Puntos Fantasy por Juego por Posición', fontsize=14, fontweight='bold')
ax4.grid(axis='y', alpha=0.3)
for i, v in enumerate(avg_fpts_per_game):
    ax4.text(i, v + 0.2, f'{v:.2f}', ha='center', fontweight='bold')

# 7.5 Top 5 jugadores de cada posición ofensiva
ax5 = fig.add_subplot(gs[2, :])
top_players = []
top_fpts = []
top_colors = []
top_positions = []

for pos, df_pos, color in [('QB', df_QB, 'purple'), ('RB', df_RB, 'orangered'), 
                            ('TE', df_TE, 'teal'), ('WR', df_WR, 'crimson')]:
    top_5 = df_pos.nlargest(5, 'FPTS')
    for _, player in top_5.iterrows():
        top_players.append(player['Player'].split('(')[0].strip()[:15])
        top_fpts.append(player['FPTS'])
        top_colors.append(color)
        top_positions.append(pos)

y_pos = np.arange(len(top_players))
ax5.barh(y_pos, top_fpts, color=top_colors, alpha=0.7, edgecolor='black')
ax5.set_yticks(y_pos)
ax5.set_yticklabels(top_players, fontsize=9)
ax5.set_xlabel('Fantasy Points', fontsize=12, fontweight='bold')
ax5.set_title('Top 5 Jugadores por Posición Ofensiva', fontsize=14, fontweight='bold')
ax5.invert_yaxis()
ax5.grid(axis='x', alpha=0.3)

# Agregar leyenda
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='purple', alpha=0.7, label='QB'),
                   Patch(facecolor='orangered', alpha=0.7, label='RB'),
                   Patch(facecolor='teal', alpha=0.7, label='TE'),
                   Patch(facecolor='crimson', alpha=0.7, label='WR')]
ax5.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.savefig('EDA_Comparativo.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Gráficas comparativas guardadas en 'EDA_Comparativo.png'")

# ============================================
# RESUMEN FINAL
# ============================================
print("\n" + "=" * 100)
print("RESUMEN FINAL DEL ANÁLISIS")
print("=" * 100)

print(f"\n📊 DATASET COMBINADO:")
print(f"   Total de jugadores/equipos: {len(df)}")
print(f"   Total de columnas: {len(df.columns)}")

print(f"\n🏈 DISTRIBUCIÓN POR POSICIÓN:")
for pos in ['DST', 'K', 'QB', 'RB', 'TE', 'WR']:
    count = len(df[df['Position'] == pos])
    print(f"   {pos}: {count} jugadores")

print(f"\n⭐ TOP PERFORMERS POR POSICIÓN:")
print(f"   DST: {df_DST.nlargest(1, 'FPTS').iloc[0]['Player']} ({df_DST.nlargest(1, 'FPTS').iloc[0]['FPTS']:.1f} pts)")
print(f"   K:   {df_K.nlargest(1, 'FPTS').iloc[0]['Player']} ({df_K.nlargest(1, 'FPTS').iloc[0]['FPTS']:.1f} pts)")
print(f"   QB:  {df_QB.nlargest(1, 'FPTS').iloc[0]['Player']} ({df_QB.nlargest(1, 'FPTS').iloc[0]['FPTS']:.1f} pts)")
print(f"   RB:  {df_RB.nlargest(1, 'FPTS').iloc[0]['Player']} ({df_RB.nlargest(1, 'FPTS').iloc[0]['FPTS']:.1f} pts)")
print(f"   TE:  {df_TE.nlargest(1, 'FPTS').iloc[0]['Player']} ({df_TE.nlargest(1, 'FPTS').iloc[0]['FPTS']:.1f} pts)")
print(f"   WR:  {df_WR.nlargest(1, 'FPTS').iloc[0]['Player']} ({df_WR.nlargest(1, 'FPTS').iloc[0]['FPTS']:.1f} pts)")

print(f"\n📈 PROMEDIOS DE FPTS POR POSICIÓN:")
for pos, df_pos in [('DST', df_DST), ('K', df_K), ('QB', df_QB), ('RB', df_RB), ('TE', df_TE), ('WR', df_WR)]:
    print(f"   {pos}: {df_pos['FPTS'].mean():.2f} puntos (σ={df_pos['FPTS'].std():.2f})")

print(f"\n💾 ARCHIVOS GENERADOS:")
print(f"   ✓ EDA_DST.png - Análisis de Defensas")
print(f"   ✓ EDA_Kickers.png - Análisis de Pateadores")
print(f"   ✓ EDA_QB.png - Análisis de Quarterbacks")
print(f"   ✓ EDA_RB.png - Análisis de Running Backs")
print(f"   ✓ EDA_TE.png - Análisis de Tight Ends")
print(f"   ✓ EDA_WR.png - Análisis de Wide Receivers")
print(f"   ✓ EDA_Comparativo.png - Análisis Comparativo entre Posiciones")

print("\n" + "=" * 100)
print("✅ ANÁLISIS EXPLORATORIO COMPLETADO")
print("=" * 100)
```

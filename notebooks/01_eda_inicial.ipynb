```python
import pandas as pd

df_DST = pd.read_csv("../data/FantasyPros_Fantasy_Football_Statistics_DST.csv")
df_K = pd.read_csv("../data/FantasyPros_Fantasy_Football_Statistics_K.csv")
df_QB = pd.read_csv("../data/FantasyPros_Fantasy_Football_Statistics_QB.csv")
df_RB = pd.read_csv("../data/FantasyPros_Fantasy_Football_Statistics_RB.csv")
df_TE = pd.read_csv("../data/FantasyPros_Fantasy_Football_Statistics_TE.csv")
df_WR = pd.read_csv("../data/FantasyPros_Fantasy_Football_Statistics_WR.csv")

df_DST['Position'] = 'DST'
df_K['Position'] = 'K'
df_QB['Position'] = 'QB'
df_RB['Position'] = 'RB'
df_TE['Position'] = 'TE'
df_WR['Position'] = 'WR'

df = pd.concat([df_DST, df_K, df_QB, df_RB, df_TE, df_WR], ignore_index=True)

print(f"Total de filas: {len(df)}")
print(f"\nDistribución por posición:")
print(df['Position'].value_counts())
print(f"\nPrimeras filas:")
print(df.head())
print(f"\nColumnas:")
print(df.columns.tolist())
print(f"\nInfo del DataFrame:")
print(df.info())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# Configurar estilo
sns.set_style("whitegrid")
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (20, 12)
plt.rcParams['font.size'] = 10

print("=" * 100)
print("ANÁLISIS EXPLORATORIO DE DATOS - FANTASY FOOTBALL 2024")
print("=" * 100)

# ============================================
# 1. ANÁLISIS DE DEFENSAS (DST)
# ============================================
print("\n" + "=" * 100)
print("1. ANÁLISIS DE DEFENSAS (DST)")
print("=" * 100)

print(f"\nTotal de equipos: {len(df_DST)}")
print(f"\nEstadísticas descriptivas:")
print(df_DST.describe())

print(f"\n--- Top 10 Defensas por FPTS ---")
top_dst = df_DST.nlargest(10, 'FPTS')
print(top_dst[['Rank', 'Player', 'SACK', 'INT', 'FR', 'DEF TD', 'SPC TD', 'FPTS', 'FPTS/G']])

# Gráficas DST
fig = plt.figure(figsize=(20, 12))
gs = GridSpec(3, 3, figure=fig)

# 1.1 Top 10 DST por FPTS
ax1 = fig.add_subplot(gs[0, :2])
top_dst_plot = df_DST.nlargest(10, 'FPTS')
ax1.barh(top_dst_plot['Player'], top_dst_plot['FPTS'], color='darkblue')
ax1.set_xlabel('Fantasy Points', fontsize=12, fontweight='bold')
ax1.set_title('Top 10 Defensas por Puntos Fantasy', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, v in enumerate(top_dst_plot['FPTS']):
    ax1.text(v + 0.5, i, f'{v:.1f}', va='center', fontweight='bold')

# 1.2 Distribución de FPTS/G
ax2 = fig.add_subplot(gs[0, 2])
ax2.hist(df_DST['FPTS/G'], bins=15, color='steelblue', edgecolor='black', alpha=0.7)
ax2.set_xlabel('FPTS/G', fontsize=10, fontweight='bold')
ax2.set_ylabel('Frecuencia', fontsize=10, fontweight='bold')
ax2.set_title('Distribución de FPTS por Juego', fontsize=12, fontweight='bold')
ax2.axvline(df_DST['FPTS/G'].mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {df_DST["FPTS/G"].mean():.2f}')
ax2.legend()

# 1.3 Correlación entre estadísticas defensivas
ax3 = fig.add_subplot(gs[1, :])
corr_cols = ['SACK', 'INT', 'FR', 'FF', 'DEF TD', 'SFTY', 'SPC TD', 'FPTS']
corr_matrix = df_DST[corr_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, ax=ax3, cbar_kws={'label': 'Correlación'})
ax3.set_title('Matriz de Correlación - Estadísticas Defensivas', fontsize=14, fontweight='bold')

# 1.4 Comparación de estadísticas defensivas (Top 10)
ax4 = fig.add_subplot(gs[2, :])
top_10_dst = df_DST.nlargest(10, 'FPTS')
x = np.arange(len(top_10_dst))
width = 0.15
ax4.bar(x - 2*width, top_10_dst['SACK'], width, label='Sacks', alpha=0.8)
ax4.bar(x - width, top_10_dst['INT'], width, label='Intercepciones', alpha=0.8)
ax4.bar(x, top_10_dst['FR'], width, label='Fumbles Rec', alpha=0.8)
ax4.bar(x + width, top_10_dst['DEF TD'], width, label='DEF TD', alpha=0.8)
ax4.bar(x + 2*width, top_10_dst['SPC TD'], width, label='SPC TD', alpha=0.8)
ax4.set_xlabel('Equipos', fontsize=12, fontweight='bold')
ax4.set_ylabel('Cantidad', fontsize=12, fontweight='bold')
ax4.set_title('Comparación de Estadísticas Defensivas - Top 10', fontsize=14, fontweight='bold')
ax4.set_xticks(x)
ax4.set_xticklabels([p.split('(')[0].strip()[:10] for p in top_10_dst['Player']], rotation=45, ha='right')
ax4.legend()
ax4.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('EDA_DST.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Gráficas de DST guardadas en 'EDA_DST.png'")
```

```python
# ============================================
# 2. ANÁLISIS DE KICKERS (K)
# ============================================
print("\n" + "=" * 100)
print("2. ANÁLISIS DE PATEADORES (K)")
print("=" * 100)

print(f"\nTotal de kickers: {len(df_K)}")
print(f"\nEstadísticas descriptivas:")
print(df_K.describe())

print(f"\n--- Top 10 Kickers por FPTS ---")
top_k = df_K.nlargest(10, 'FPTS')
print(top_k[['Rank', 'Player', 'FG', 'FGA', 'PCT', 'XPT', 'FPTS', 'FPTS/G']])

# Gráficas Kickers
fig = plt.figure(figsize=(20, 12))
gs = GridSpec(3, 3, figure=fig)

# 2.1 Top 10 Kickers por FPTS
ax1 = fig.add_subplot(gs[0, :2])
top_k_plot = df_K.nlargest(10, 'FPTS')
ax1.barh(top_k_plot['Player'], top_k_plot['FPTS'], color='darkgreen')
ax1.set_xlabel('Fantasy Points', fontsize=12, fontweight='bold')
ax1.set_title('Top 10 Kickers por Puntos Fantasy', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, v in enumerate(top_k_plot['FPTS']):
    ax1.text(v + 0.5, i, f'{v:.0f}', va='center', fontweight='bold')

# 2.2 Distribución de precisión (PCT)
ax2 = fig.add_subplot(gs[0, 2])
ax2.hist(df_K['PCT'], bins=15, color='green', edgecolor='black', alpha=0.7)
ax2.set_xlabel('Porcentaje de Precisión', fontsize=10, fontweight='bold')
ax2.set_ylabel('Frecuencia', fontsize=10, fontweight='bold')
ax2.set_title('Distribución de Precisión de Kickers', fontsize=12, fontweight='bold')
ax2.axvline(df_K['PCT'].mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {df_K["PCT"].mean():.1f}%')
ax2.legend()

# 2.3 Relación entre FG y FPTS
ax3 = fig.add_subplot(gs[1, 0])
ax3.scatter(df_K['FG'], df_K['FPTS'], alpha=0.6, s=100, color='green', edgecolors='black')
ax3.set_xlabel('Field Goals', fontsize=10, fontweight='bold')
ax3.set_ylabel('FPTS', fontsize=10, fontweight='bold')
ax3.set_title('Field Goals vs Puntos Fantasy', fontsize=12, fontweight='bold')
ax3.grid(alpha=0.3)

# 2.4 Relación entre PCT y FPTS
ax4 = fig.add_subplot(gs[1, 1])
ax4.scatter(df_K['PCT'], df_K['FPTS'], alpha=0.6, s=100, color='darkgreen', edgecolors='black')
ax4.set_xlabel('Porcentaje de Precisión', fontsize=10, fontweight='bold')
ax4.set_ylabel('FPTS', fontsize=10, fontweight='bold')
ax4.set_title('Precisión vs Puntos Fantasy', fontsize=12, fontweight='bold')
ax4.grid(alpha=0.3)

# 2.5 Distribución de goles por distancia (Top 10)
ax5 = fig.add_subplot(gs[1, 2])
top_10_k = df_K.nlargest(10, 'FPTS')
distance_data = top_10_k[['1-19', '20-29', '30-39', '40-49', '50+']].mean()
ax5.bar(distance_data.index, distance_data.values, color=['lightgreen', 'green', 'darkgreen', 'olive', 'darkolivegreen'])
ax5.set_xlabel('Distancia', fontsize=10, fontweight='bold')
ax5.set_ylabel('Promedio de FG', fontsize=10, fontweight='bold')
ax5.set_title('Promedio de FG por Distancia (Top 10)', fontsize=12, fontweight='bold')
ax5.grid(axis='y', alpha=0.3)

# 2.6 Comparación de métricas clave (Top 10)
ax6 = fig.add_subplot(gs[2, :])
x = np.arange(len(top_10_k))
width = 0.25
ax6.bar(x - width, top_10_k['FG'], width, label='FG', alpha=0.8, color='green')
ax6.bar(x, top_10_k['XPT'], width, label='XPT', alpha=0.8, color='lightgreen')
ax6.bar(x + width, top_10_k['50+'], width, label='50+ Yardas', alpha=0.8, color='darkgreen')
ax6.set_xlabel('Kickers', fontsize=12, fontweight='bold')
ax6.set_ylabel('Cantidad', fontsize=12, fontweight='bold')
ax6.set_title('Comparación de Métricas Clave - Top 10 Kickers', fontsize=14, fontweight='bold')
ax6.set_xticks(x)
ax6.set_xticklabels([p.split('(')[0].strip()[:10] for p in top_10_k['Player']], rotation=45, ha='right')
ax6.legend()
ax6.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('EDA_Kickers.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Gráficas de Kickers guardadas en 'EDA_Kickers.png'")
```
